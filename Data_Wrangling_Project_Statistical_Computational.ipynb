# Code 02

```python
# Code Untuk Soal No 02

nd_df = pd.merge(dosen, nilai, on='NIDN')
mean = nd_df[['Pendidikan_Terakhir', 'nilai' ]]
mean = mean.groupby(nd_df['Pendidikan_Terakhir']).mean()[['nilai']]
mean = mean.rename({'nilai': 'mean_nilai'}, axis=1)

std = nd_df.groupby(nd_df['Pendidikan_Terakhir']).std()[['nilai']]
std = std.rename({'nilai': 'std_nilai'}, axis=1)

nilaipt_dosen = pd.concat([mean, std], axis=1)
nilaipt_dosen
```

# Interpretasi 02:

* Terlihat pada data bahwa rata - rata pemberian nilai dosen dengan pendidikan terakhir S3 lebih tinggi daripada dosen dengan pendidikan terakhir S2. Rata-rata pemberian nilai dosen dengan pendidikan terakhir S3 sebesar 83.266667, sedangkan rata-rata pemberian nilai dosen dengan pendidikan terakhir S2 sebesar 64.166667. Pengaruh std_nilai terhadap mean_nilai adalah semakin rendah nilai std, semakin dekat mendekati rata-rata (tidak tersebar). Sementara itu, jika nilai std tinggi, akan semakin lebar rentang variasi datanya (semakin tersebar).

- Rata-rata pemberian nilai dosen dengan pendidikan terakhir S3, lebih mendekati nilai rata-rata dengan std yang kecil (12)
- Rata-rata pemberian nilai dosen dengan pendidikan terakhir S2, memiliki variasi data yang tersebar dengan std yang tinggi (15)
- Besar nilai standar deviasi merepresentasikan pesebaran data. Semakin besar nilai std-nya maka semakin beragam nilainya atau semakin tidak akurat nilai mean nya

Kesimpulannya, terdapat perbedaan yang signifikan dalam pemberian nilai antara dosen dengan pendidikan terakhir S3 dan S2.

# Code 06 #2

```python
# Code soal No 06 #2

nilai_m = nilai[['id_mata_kuliah','nilai']]
mean = nilai_m.groupby('id_mata_kuliah')[['nilai']].agg('mean')
mean = mean.rename({'nilai': 'mean_nilai'}, axis=1)

std = nilai_m.groupby(nilai_m['id_mata_kuliah']).std()[['nilai']]
std = std.rename({'nilai': 'std_nilai'}, axis=1)
left = pd.concat([mean,std],axis = 1)

matkul = pd.DataFrame(nilai['id_mata_kuliah'].value_counts())
right = matkul.rename({'id_mata_kuliah': 'frekuensi_mahasiswa_per_kelas'}, axis=1)
right.reset_index(inplace=True)
right = right.rename(columns = {'index':'matkul'})

matakuliah = pd.merge(left, right, left_on = 'id_mata_kuliah', right_on = 'matkul')
matakuliah = matakuliah.set_index('matkul')
matakuliah
```

# Interpretasi 06 #2

* Terdapat informasi lain dalam data, yaitu berdasarkan data wrangling di atas, didapat mata kuliah dengan jumlah mahasiswa di kelas mata kuliah tersebut tidak berpengaruh terhadap rata-rata nilai mahasiswa tsb. Dengan kelas KAL01 yang memiliki jumlah mahasiswa di kelas paling banyak, yaitu 15, memiliki rata-rata nilai yaitu 78.666667 (dilihat dari standar deviasinya, maka mean tepat dijadikan acuan sebagai nilai tengah data).
