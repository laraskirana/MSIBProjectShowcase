**Pre-Processing**




```python
def loadTweets(file='Tweets.json'):
    f=open(file,encoding='utf-8', errors ='ignore', mode='r')
    T=f.readlines();f.close()
    for i,t in enumerate(T):
        T[i] = json.loads(t.strip())
    return T
```

Selanjutnya, melakukan ***URL Analysis and Cleaning*** dengan tujuan menghilangkan URL dan HTTPS dalam tweets.

```python
import re
# Website URLS http(s) .... untuk ftp trivial
urlPattern = re.compile(r't.co/(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')

T5 = ' '.join([str(item) for item in T4])
T6=re.sub(urlPattern,'',T5)
T6 = T6.split()
print(T6)
```

Selanjutnya, melakukan ***Emoji Analysis dan Cleaning*** dengan tujuan menghilangkan emoji yang ada di tweets.

```python

import re
def remove_emojis(T6):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642"
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', T6)
```
